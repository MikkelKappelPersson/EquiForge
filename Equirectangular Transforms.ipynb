## Parameter Reference

- **FOV**: Standard DSLR (60-90°), Wide-angle (100-120°), Smartphone (70-80°)
- **Yaw**: Left/right rotation (0° = forward)
- **Pitch**: Up/down rotation (0° = level, positive = up)
- **Roll**: Rotation around depth axis (clockwise/counterclockwise) in degrees

```python
# Import the optimized function from the new package structure
import time
import matplotlib.pyplot as plt

from equiforge import pers2equi


start_time = time.time()

# Run optimized conversion (will use multiple CPU cores or GPU if available)
optimized_result = pers2equi(
    'Images/SamplePerspectiveImage.jpg',
    output_height=4096,
    fov_h=90.0,
    yaw=0.0,
    pitch=0.0,
    roll=0.0
)

end_time = time.time()

# Display the result
plt.figure(figsize=(16, 8))
plt.imshow(optimized_result)
plt.title(f"Optimized conversion (completed in {end_time - start_time:.2f} seconds)")
plt.axis('off')
plt.show()

print(f"Processing completed in {end_time - start_time:.2f} seconds")
```

## Equirectangular to Perspective Conversion

Convert an equirectangular image back to perspective view:
- The parameters define where in the equirectangular image to "place the camera"
- Adjust yaw to look left/right, pitch to look up/down

```python
from equiforge import equi2pers
import time

# Use previously created equirectangular image as input
equi_image = "Images/EquiSample.jpg"  # Or load your own: np.array(Image.open('path/to/equirect.jpg'))

start_time = time.time()

# Convert back to perspective from desired viewing direction
pers_result = equi2pers(
    equi_image,
    output_width=1920,
    output_height=1080,
    fov_h=90.0,
    yaw=0.0,    # Look 45° to the right
    pitch=0.0,  # Look 15° up
    roll=0.0
)

end_time = time.time()

# Display the result
plt.figure(figsize=(16, 9))
plt.imshow(pers_result)
plt.title(f"Perspective view (yaw=45°, pitch=15°, completed in {end_time - start_time:.2f} seconds)")
plt.axis('off')
plt.show()

print(f"Processing completed in {end_time - start_time:.2f} seconds")
```
