## Parameter Reference

- **FOV**: Standard DSLR (60-90°), Wide-angle (100-120°), Smartphone (70-80°)
- **Yaw**: Left/right rotation (0° = forward)
- **Pitch**: Up/down rotation (0° = level, positive = up)
- **Roll**: Clockwise/counterclockwise rotation

```python
# Import the optimized function from the new package structure
import time
import matplotlib.pyplot as plt

from equiforge import pers2equi


start_time = time.time()

# Run optimized conversion (will use multiple CPU cores or GPU if available)
optimized_result = pers2equi(
    'Images/SamplePerspectiveImage.jpg',
    output_height=4096,
    fov_h=90.0,
    yaw=0.0,
    pitch=0.0,
    roll=0.0
)

end_time = time.time()

# Display the result
plt.figure(figsize=(16, 8))
plt.imshow(optimized_result)
plt.title(f"Optimized conversion (completed in {end_time - start_time:.2f} seconds)")
plt.axis('off')
plt.show()

print(f"Processing completed in {end_time - start_time:.2f} seconds")
```
